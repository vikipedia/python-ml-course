````python
def cone():
    tx = Translate(x=30, y=0) | Scale(sx=0.8,sy=0.8)
    shape = Circle(center=Point(x=-50, y=0), stroke='blue') | Repeat(n=18, transformation=tx)
    return shape
````

````python
cone()
````

````python
tx = Rotate(angle=5) | Scale(sx=0.92,sy=0.92)
Rectangle(width=200, height=200) | Repeat(n=360//5, transformation=tx)
````

````python
def square_design(size, n, scale):
    tx = Rotate(angle=360//n) | Scale(sx=scale,sy=scale)
    return Rectangle(width=size, height=size) | Repeat(n=n, transformation=tx)
````

````python
square_design(300, 40,0.87)
````

**Problems**
- Write a function `tile` which can tile given shape in m rows and n columns

````python
def tile(shape, size,m,n):
    neworigin = size/2-150
    shape = shape | Translate(x=neworigin, y=neworigin)
    xshape = shape | Repeat(n=n, transformation=Translate(x=size, y=0))
    return xshape | Repeat(n=m, transformation=Translate(x=0,y=size))
````

````python
t = tile(square_design(30, 40, 0.87), 30,10,10)
t
````
